```python
# imports here
import matplotlib.pyplot as plt
import numpy as np
# variables
x = np.arange(0,11)
y1=2*x+1
y2=x**2

# plotting
plt.plot(x,y1, label='y1=2x+1')
plt.plot(x,y2, label='y2=x**2')
plt.legend()
# labelling
plt.xlabel('x')
plt.ylabel('y')
plt.title('graph of y=2x+1')

# grid
plt.grid(True)
# show
plt.show()
```
