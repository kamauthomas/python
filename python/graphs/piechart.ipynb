```python
# imports here
import numpy as np
import matplotlib.pyplot as plt
```

```python
# variables
categories = ["Apples", "Bananas", "Cherries", "Dates"]
values = np.random.randint(10, 50, size = 4)

# plotting
plt.pie(values, labels = categories, autopct = '%1.1f%%')
plt.title("Fruit Distribution")
plt.legend()
plt.show()
```
